```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Import Data Set by going in file in LHS 
df= pd.read_csv("cubic_zirconia.csv")
df.head() #Displays random samples
```

```python
df.info() #check for data tyes
```

```python
cols = [col for col in df.columns if
           df[col].dtype == "object"]
cols
```

 Exploring object columns 

```python
df['cut'].value_counts().plot.bar()
```

```python
df['color'].value_counts().plot.bar()
```

```python
df['clarity'].value_counts().plot.bar()
```

Numerical Values

```python
num_cols = [col for col in df.columns if
           df[col].dtype == 'int' or df[col].dtype == 'float']
```

```python
plt.figure(figsize = (11,8))

for i,col in enumerate(num_cols):
    plt.subplot(3,3,i+1)
    sns.histplot( x = col, data = df, 
                 bins = 30, color = 'steelblue')
    plt.tight_layout()
```

Bivariate Analysis 

```python
#corelating carats and price 
df[df['price'] < 1000].sample(1000).plot.scatter(x='price', y='carat')
```

```python
#corelating carats and price 
df[df['price'] < 1000].plot.hexbin(x='price', y='carat', gridsize=15)
```

```python
sns.heatmap(df.corr(), annot = True, robust = True, cmap= 'coolwarm')
```
